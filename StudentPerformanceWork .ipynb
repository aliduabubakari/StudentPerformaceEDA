```python
#Confidence interval calculation for math score 

import scipy.stats as stats
import math 

np.random.seed(10)

#take 100 samples from the dataset to calculate the sample mean 
sample_marks = np.random.choice(a = student_data['math score'], size = 1000)
sample_mean = sample_marks.mean()

#Get the Z-critical value 
z_critical = stats.norm.ppf(q=0.95)

#Check the Z-critical value
print()
print('The z-critical for maths score: ', z_critical)

#Get the population standard deviation value
pop_stdev = student_data['math score'].std()

#Check the population standard deviation value
print()
print('The population standard deviation for maths score: ', pop_stdev)

#Check the margin of error
margin_error = z_critical*(pop_stdev/math.sqrt(1000))

#Dtermine the the confidence interval 
confidence_interval =  (sample_mean - margin_error, sample_mean + margin_error) 
print()
print('the confidence interval for math score: ', confidence_interval)
print()
print('The True Mean for math score: ', student_data['math score'].mean())
```

```python
#Assigning gardes to the exam scores according to the following rules

#0- 40 marks: grade E
#41- 60 marks: grade D
#61- 70 marks: grade C
#71- 80 marks: grade B
#81- 90 marks: grade A
#91- 100 marks: grade O

def getgrade(percentage, status):
    if status == 'Fail':
        return 'E'
    if (91 <= percentage and percentage <= 100):
        return 'O'
    if (81 <= percentage and percentage <= 90):
        return 'A'
    if (71 <= percentage and percentage <= 80):
        return 'B'
    if (61 <= percentage and percentage <= 70):
        return 'C'
    if (41 <= percentage and percentage <= 60):
        return 'D'
    else:
        return 'E'
    
```

```python
student_data['grades'] = student_data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1)

student_data['grades'].value_counts()

```

### Splitting the dependendent and independent variables


```python
x = student_data.iloc[:,:14 ]
y = student_data.iloc[:,14 ]

print(x.shape)
print(y.shape)
```

```python
#splitting the dataset into taining and test sets 

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 45)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#importing the minmaxscaler 

from sklearn.preprocessing import MinMaxScaler

#create a scaler model 
scaler = MinMaxScaler()

#input the training data into the minmaxscaler  
x_train = scaler.fit_transform(x_train)

#input the test data into the minmaxscaler  
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

#create a model 
Random_model = RandomForestClassifier()

#input the training data into the model 
Random_model.fit(x_train, y_train)

#predicting the x-test results 
y_pred = Random_model.predict(x_test)

#Calculating the accuracies 

print('Training Accuracy:', Random_model.score(x_train, y_train))
print('Testing Accuracy:', Random_model.score(x_test, y_test))
```

```python
#printing the confusion matrix 

plt.subplots(figsize = (15,8))

from sklearn.metrics import confusion_matrix

#creating the confusion matrix 
confusionMatrix = confusion_matrix(y_test, y_pred)

#priting the confusion matrix 
plt.rcParams['figure.figsize'] = (5,5)
sns.heatmap(confusionMatrix, annot = True, cmap ='Purples')
```

```python
#Classification Report

from sklearn.metrics import classification_report 

Class_Report = classification_report(y_test, y_pred)
print(Class_Report)
```
